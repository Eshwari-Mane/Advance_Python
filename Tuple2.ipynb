- Mutability: List are Mutable  . Tuples are Immutable
- Time:List are the time consuming where as Tuple is faster then list
- Methods : List have methods tuple dont have
- Memory : List consume more memeories.Tuple consume less memory as compared to list
- Operations:
- Error : In List their are more chances of occuring the error.In Tuple their are less chances of ocurring the error.
