- Unchangeable

```python
#Len()
print(len(my_tuple))
```

```python
#create Tuple with one item: To create this type of tuple we have to add one semicon after the element.Otherwise it is not considered as a tuple
city1 = ("chennai",)
print(type(city1))
```

```python
#Tuple constructor
city = tuple(("Mumbai","Pune","nagpur","goa"))
print(city)
```
